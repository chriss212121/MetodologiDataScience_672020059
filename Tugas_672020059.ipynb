Soal 2: Permen

```python
import matplotlib.pyplot as plt

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
Jumlah_permen = [52, 39, 78, 13, 78]
warna = ('#5DADE2', '#8B4513', '#FF4081', '#FFD700', '#FFA500')
peluang_kopiko = (0,0.1,0,0,0)

plt.title('Peluang Permen')
plt.pie(
    Jumlah_permen,
    labels=nama_permen,
    autopct='%1.2f%%',
    colors = warna,
    explode = peluang_kopiko,
    shadow = True

)

plt.show()
```

Soal 3: Makanan

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

makanan = ('Donat', 'Pastel', 'Kue Coklat', 'Ice Cream', 'Puding Vanila', 'Brownies', 'Puding Strawberi', 'Puding Coklat', 'Ice Cream Nutela', 'Kue Coklat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
jumlah = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({
    'Mkn':makanan,
    'Jum':jumlah,
})
df.sort_values(by='Jum', inplace=True)

x_coord = np.arange(len(df))
plt.bar(x_coord, df['Jum'], tick_label=df['Mkn'])
plt.xticks(rotation = 90)
plt.ylabel('Jumlah')
plt.title('Nama Makanan Penutup')
plt.show()
```

Tiga makanan penutup yang harus
disingkirkan adalah Kue wajik, Pastel, Puding Vanila
